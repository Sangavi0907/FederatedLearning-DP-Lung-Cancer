```python
import os
import cv2
import numpy as np

data_path = '/content/drive/MyDrive/DATASET/Data'

img_size = 128

def load_images_from_folder(folder_path):
    images = []
    labels = []
    class_names = os.listdir(folder_path)

    for class_name in class_names:
        class_path = os.path.join(folder_path, class_name)
        if not os.path.isdir(class_path):
            continue

        for img_name in os.listdir(class_path):
            img_path = os.path.join(class_path, img_name)
            img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
            if img is not None:
                img = cv2.resize(img, (img_size, img_size))
                img = np.expand_dims(img, axis=-1)
                images.append(img)
                labels.append(class_name)

    return np.array(images), np.array(labels)


X_train, y_train = load_images_from_folder(os.path.join(data_path, 'train'))
X_test, y_test = load_images_from_folder(os.path.join(data_path, 'test'))
X_valid, y_valid = load_images_from_folder(os.path.join(data_path, 'valid'))


X_train = X_train / 255.0
X_test = X_test / 255.0
X_valid = X_valid / 255.0


print(f"Train images shape: {X_train.shape}, Train labels shape: {y_train.shape}")
print(f"Test images shape: {X_test.shape}, Test labels shape: {y_test.shape}")
print(f"Valid images shape: {X_valid.shape}, Valid labels shape: {y_valid.shape}")
```

# **GAN-Based Synthetic Data Generation**

```python
import os, numpy as np, cv2, matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tqdm import tqdm

def build_generator():
    model = keras.Sequential([
        layers.Dense(16*16*128, activation="relu", input_dim=100),
        layers.Reshape((16, 16, 128)),
        layers.UpSampling2D(),
        layers.Conv2D(128, kernel_size=3, padding="same"),
        layers.BatchNormalization(), layers.ReLU(),
        layers.UpSampling2D(),
        layers.Conv2D(64, kernel_size=3, padding="same"),
        layers.BatchNormalization(), layers.ReLU(),
        layers.Conv2D(1, kernel_size=3, padding="same", activation="sigmoid")
    ])
    return model

def build_discriminator():
    model = keras.Sequential([
        layers.Conv2D(64, kernel_size=3, strides=2, padding="same", input_shape=(64, 64, 1)),
        layers.LeakyReLU(0.2), layers.Dropout(0.3),
        layers.Conv2D(128, kernel_size=3, strides=2, padding="same"),
        layers.LeakyReLU(0.2), layers.Dropout(0.3),
        layers.Flatten(), layers.Dense(1, activation='sigmoid')
    ])
    return model

def build_gan(generator, discriminator):
    discriminator.trainable = False
    gan_input = keras.Input(shape=(100,))
    fake_img = generator(gan_input)
    gan_output = discriminator(fake_img)
    model = keras.Model(gan_input, gan_output)
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model

generator = build_generator()
discriminator = build_discriminator()
discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
gan = build_gan(generator, discriminator)

# Prepare training data (grayscale images)
real_images_array = np.array([cv2.resize(img.squeeze(), (64, 64)) for img in X_full])
real_images_array = np.expand_dims(real_images_array, axis=-1).astype("float32") / 255.0

# GAN Training Loop
epochs = 1000
batch_size = 32
noise_dim = 100

for epoch in tqdm(range(epochs + 1)):
    idx = np.random.randint(0, real_images_array.shape[0], batch_size)
    real_imgs = real_images_array[idx]
    noise = np.random.normal(0, 1, (batch_size, noise_dim))
    fake_imgs = generator.predict(noise)
    real_labels, fake_labels = np.ones((batch_size, 1)), np.zeros((batch_size, 1))
    d_loss_real = discriminator.train_on_batch(real_imgs, real_labels)
    d_loss_fake = discriminator.train_on_batch(fake_imgs, fake_labels)
    noise = np.random.normal(0, 1, (batch_size, noise_dim))
    gan.train_on_batch(noise, np.ones((batch_size, 1)))

# Generate synthetic images
num_synthetic = 200
random_noise = np.random.normal(0, 1, (num_synthetic, 100))
synthetic_images = generator.predict(random_noise)
synthetic_images_resized = np.array([cv2.resize(img.squeeze(), (128, 128)) for img in synthetic_images])
synthetic_images_resized = np.expand_dims(synthetic_images_resized, axis=-1) / 255.0

# Assign all generated images as CANCER class (1)
synthetic_labels = np.ones(synthetic_images_resized.shape[0])  # Class 1 = Cancer
```

```python
synthetic_images_scaled = (synthetic_images * 255).astype(np.uint8)
np.save("synthetic_images.npy", synthetic_images_scaled)
```

```python
# Merge with real dataset (now with predicted labels)
X_imgs_total = np.concatenate([X_full, synthetic_images_resized], axis=0)
y_total = np.concatenate([y_full, synthetic_labels], axis=0)

# Optional Shuffle
indices = np.arange(X_imgs_total.shape[0])
np.random.shuffle(indices)

X_imgs_total = X_imgs_total[indices]
y_total = y_total[indices]
```

**Split Function for Federated Setup**

```python
def split_data_for_fog_nodes(X_img, X_meta, y, num_nodes):
    min_len = min(len(X_img), len(X_meta), len(y))
    div_len = (min_len // num_nodes) * num_nodes

    X_img = X_img[:div_len]
    X_meta = X_meta[:div_len]
    y = y[:div_len]

    X_img_splits = np.array_split(X_img, num_nodes)
    X_meta_splits = np.array_split(X_meta, num_nodes)
    y_splits = np.array_split(y, num_nodes)

    for i in range(num_nodes):
        assert len(X_img_splits[i]) == len(X_meta_splits[i]) == len(y_splits[i])
    return X_img_splits, X_meta_splits, y_splits
```

**Multimodal Model (CNN + FFNN)**

```python
from tensorflow.keras import layers, Input, Model

def build_cnn_ffnn_binary(image_shape, metadata_shape):
    img_input = Input(shape=image_shape)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(img_input)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)

    meta_input = Input(shape=(metadata_shape,))
    m = layers.Dense(64, activation='relu')(meta_input)
    m = layers.Dense(32, activation='relu')(m)

    combined = layers.concatenate([x, m])
    z = layers.Dense(64, activation='relu')(combined)
    output = layers.Dense(1, activation='sigmoid')(z)  # Binary classification

    model = Model(inputs=[img_input, meta_input], outputs=output)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

```

**Privacy Algorithms (Generalization + Threshold Shuffler)**

```python
def generalization_algorithm(local_weights):
    processed_weights = []
    for weight in local_weights:
        noise = np.random.normal(0, 0.01, size=weight.shape)
        processed_weights.append(weight + noise)
    return processed_weights

def threshold_shuffler_algorithm(weights):
    shuffled_weights = []
    for W in weights:
        l_initial = 0.9 * W
        l_final = 1.1 * W
        shuffled_weights.append(np.random.uniform(l_initial, l_final))
    return shuffled_weights
```

**Fog Node Training Function**

```python
import boto3, pickle, os

def add_adversarial_noise(X, epsilon=0.01):
    noise = epsilon * np.sign(np.random.randn(*X.shape))
    return np.clip(X + noise, 0, 1)

def train_fog_node(X_img, X_meta, y, node_id, num_nodes, model_builder, s3_client, BUCKET_NAME):
    print(f"\nüîÑ Fog Node {node_id+1}/{num_nodes} Training Started...")
    X_img_noisy = add_adversarial_noise(X_img)

    model = model_builder(
        image_shape=(128, 128, 1),
        metadata_shape=X_meta.shape[1]
    )

    model.fit([X_img_noisy, X_meta], y, epochs=5, batch_size=32, verbose=1)

    weights = model.get_weights()
    generalized = generalization_algorithm(weights)
    processed = threshold_shuffler_algorithm(generalized)

    weight_file = f"fog_node_{num_nodes}_{node_id+1}.pkl"
    with open(weight_file, "wb") as f:
        pickle.dump(processed, f)
    s3_client.upload_file(weight_file, BUCKET_NAME, weight_file)
    os.remove(weight_file)
    print(f"Uploaded {weight_file} to AWS S3.")
```

**Aggregating Weights from AWS S3 & Upload Final Model**

```python
import pickle
import boto3
import os
import numpy as np

S3_BUCKET = "fedarated-learning"
s3_client = boto3.client("s3")

def aggregate_weights_from_s3(num_nodes):
    print(f"\nAggregating weights from {num_nodes} fog nodes...\n")
    aggregated_weights = None

    for i in range(num_nodes):
        file_name = f"fog_node_{num_nodes}_{i+1}.pkl"
        local_path = f"/tmp/{file_name}"

        print(f" Downloading {file_name} from S3...")
        s3_client.download_file(S3_BUCKET, file_name, local_path)

        with open(local_path, "rb") as f:
            weights = pickle.load(f)

        if not isinstance(weights, list):
            raise ValueError(f" Unexpected format in {file_name}. Expected list of numpy arrays.")

        if aggregated_weights is None:
            aggregated_weights = [np.copy(w) for w in weights]
        else:
            for j in range(len(weights)):
                if aggregated_weights[j].shape != weights[j].shape:
                    raise ValueError(f" Weight shape mismatch in node {i+1}")
                aggregated_weights[j] += weights[j]

        os.remove(local_path)

    #  Average the weights
    aggregated_weights = [w / num_nodes for w in aggregated_weights]
    print(f" Aggregation complete. Total layers: {len(aggregated_weights)}")
    return aggregated_weights
```

**Train for Each Configuration & Upload Final Global Model to S3**

```python
fog_node_configs = [3,5,7]

for num_nodes in fog_node_configs:
    print(f"\n Federated Training with {num_nodes} Fog Nodes...")

    # Split data equally across fog nodes
    X_img_splits, X_meta_splits, y_splits = split_data_for_fog_nodes(
        X_imgs_total, X_metacsv, y_total, num_nodes
    )

    # Split Y (labels) without encoding ‚Äî already binary
    y_split = [y.astype(int) for y in y_splits]

    # Train each fog node and upload weights
    for i in range(num_nodes):
        train_fog_node(
            X_img=X_img_splits[i],
            X_meta=X_meta_splits[i],
            y=y_splits[i],
            node_id=i,
            num_nodes=num_nodes,
            model_builder=build_cnn_ffnn_binary,
            s3_client=s3_client,
            BUCKET_NAME=S3_BUCKET
        )

    #  Aggregate weights
    final_global_weights = aggregate_weights_from_s3(num_nodes)

    # Save and upload final global model
    final_model_file = f"final_global_model_{num_nodes}.pkl"
    with open(final_model_file, "wb") as f:
        pickle.dump(final_global_weights, f)

    s3_client.upload_file(final_model_file, S3_BUCKET, final_model_file)
    os.remove(final_model_file)

    print(f"Final Global Model for {num_nodes} Nodes Uploaded to S3\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras import layers, Model, Input
from sklearn.metrics import roc_auc_score, log_loss, normalized_mutual_info_score
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

#  1. Build CNN + FFNN Model for Binary Classification
def build_cnn_ffnn_binary(image_shape, metadata_shape):
    img_input = Input(shape=image_shape)
    x = layers.Conv2D(32, (3,3), activation='relu', padding='same')(img_input)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Conv2D(64, (3,3), activation='relu', padding='same')(x)
    x = layers.MaxPooling2D((2,2))(x)
    x = layers.Flatten()(x)
    x = layers.Dense(128, activation='relu')(x)

    meta_input = Input(shape=(metadata_shape,))
    m = layers.Dense(64, activation='relu')(meta_input)
    m = layers.Dense(32, activation='relu')(m)

    combined = layers.concatenate([x, m])
    z = layers.Dense(64, activation='relu')(combined)
    output = layers.Dense(1, activation='sigmoid')(z)

    model = Model(inputs=[img_input, meta_input], outputs=output)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

# ‚úÖ 2. Adversarial Noise Function
def add_adversarial_noise(X, epsilon=0.01):
    noise = epsilon * np.sign(np.random.randn(*X.shape))
    return np.clip(X + noise, 0, 1)

# ‚úÖ 3. Encode Labels
le = LabelEncoder()
y_full_encoded = le.fit_transform(y_full)
y_test_encoded = le.transform(y_test)

# ‚úÖ 4. Align Metadata with Image Dataset
metadata_df = pd.read_csv("lung_cancer_prepocessed.csv")
metadata_full = metadata_df.drop(columns='LUNG_CANCER').to_numpy() # Drop the target column 'LUNG_CANCER'

# Ensure length match
#  ‚úÖ FIX: Align metadata_full with X_full (instead of X_img_test)
if len(metadata_full) < len(X_full):  # Compare with X_full, not X_img_test
    repeats = len(X_full) // len(metadata_full) + 1
    metadata_full = np.tile(metadata_full, (repeats, 1))
metadata_full = metadata_full[:len(X_full)]  # Slice to match X_full

#  ‚úÖ FIX: Align X_meta_test, y_test_encoded lengths
min_len = min(len(X_img_test), len(X_meta_test), len(y_test_encoded))  # Added y_test_encoded
X_img_test = X_img_test[:min_len]
X_meta_test = X_meta_test[:min_len]
y_test_encoded = y_test_encoded[:min_len]  # Align y_test_encoded

# ‚úÖ 5. Train Model WITHOUT Adversarial Training
model_clean = build_cnn_ffnn_binary((128, 128, 1), X_meta_test.shape[1])
model_clean.fit([X_full, metadata_full], y_full_encoded, epochs=10, batch_size=32, verbose=0)

loss_clean, acc_clean = model_clean.evaluate([X_img_test, X_meta_test], y_test_encoded, verbose=0)
y_probs_clean = model_clean.predict([X_img_test, X_meta_test])
y_preds_clean = (y_probs_clean > 0.5).astype(int)

# ‚úÖ 6. Train Model WITH Adversarial Training
X_full_noisy = add_adversarial_noise(X_full)

model_adv = build_cnn_ffnn_binary((128, 128, 1), X_meta_test.shape[1])
model_adv.fit([X_full_noisy, metadata_full], y_full_encoded, epochs=10, batch_size=32, verbose=0)

loss_adv, acc_adv = model_adv.evaluate([X_img_test, X_meta_test], y_test_encoded, verbose=0)
y_probs_adv = model_adv.predict([X_img_test, X_meta_test])
y_preds_adv = (y_probs_adv > 0.5).astype(int)

# ‚úÖ 7. AUC Metrics
auc_clean = roc_auc_score(y_test_encoded, y_probs_clean)
auc_adv = roc_auc_score(y_test_encoded, y_probs_adv)
```
